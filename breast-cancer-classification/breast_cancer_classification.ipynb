# Preprocessing & Modeling Pipeline

```python
# ModelManager class to run multiple models and selector combinations
class ModelManager(BaseEstimator, ClassifierMixin):
    def __init__(self, models, feature_selector=None):
        """
        Initializes the ModelManager with models and an optional feature selector.
        """
        self.models = models
        self.feature_selector = feature_selector
        self.pipeline = self._create_pipeline(feature_selector)
        self.fitted_models = {}

    def _create_pipeline(self, feature_selector):
        """
        Creates the preprocessing pipeline.
        """
        steps = [
            ("imputer", IterativeImputer(random_state=42)),
            ("scaler", StandardScaler())
        ]

        if feature_selector:
            steps.append(("feature_selector", feature_selector))
        else:
            steps.append(("passthrough", "passthrough"))

        preprocessing_pipeline = Pipeline(steps)
        return preprocessing_pipeline

    def fit(self, X, y):
        """
        Fits the pipeline and models to the data.
        """
        X_transformed = self.pipeline.fit_transform(X, y)
        input_dim = X_transformed.shape[1]

        for name, model in self.models.items():
            if isinstance(model, KerasClassifier):
                updated_model = KerasClassifier(model=create_keras_model, model__input_dim=input_dim, epochs=100, verbose=0)
                self.fitted_models[name] = updated_model.fit(X_transformed, y, class_weight=class_weight_dict)

            elif isinstance(model, NeuralNetBinaryClassifier):
                model.set_params(module__num_features=input_dim)
                y_transformed = check_array(y, ensure_2d=False, dtype=np.float32)
                self.fitted_models[name] = clone(model).fit(X_transformed, y_transformed)

            elif isinstance(model, LogisticGAM):
                self.fitted_models[name] = self._configure_and_fit_logistic_gam(X_transformed, y)

            else:
                self.fitted_models[name] = clone(model).fit(X_transformed, y)

            feature_selector_name = "with " + type(self.feature_selector).__name__ if self.feature_selector else "without feature selection"
            print(f"Fitted {name} {feature_selector_name} and input_dim={input_dim if 'Neural Network' in name else 'N/A'}")

    def _configure_and_fit_logistic_gam(self, X, y, lam_value=1e2):
        """
        Dynamically create terms for Logistic GAM.
        """
        terms = TermList()
        for i in range(X.shape[1]):
            terms += s(i, lam=lam_value)

        gam = LogisticGAM(terms=terms, max_iter=1000).gridsearch(X, y)
        return gam

    def predict(self, X):
        """
        Predicts using the fitted models.
        """
        X_transformed = self.pipeline.transform(X)
        predictions = {name: model.predict(X_transformed) for name, model in self.fitted_models.items()}
        return predictions

    def evaluate(self, X, y):
        """
        Evaluates the fitted models.
        """
        X_transformed = self.pipeline.transform(X)
        scores = {}
        for name, model in self.fitted_models.items():
            y_pred = model.predict(X_transformed)
            scores[name] = self._calculate_metrics(y, y_pred)
        return scores

    def _calculate_metrics(self, y_true, y_pred):
        """
        Calculates evaluation metrics for predictions.
        """
        metrics_dict = {
            "Accuracy": accuracy_score(y_true, y_pred),
            "Precision": precision_score(y_true, y_pred, average="binary", zero_division=0),
            "Recall": recall_score(y_true, y_pred, average="binary"),
            "Specificity": self._calculate_specificity(y_true, y_pred),
            "F1 Score": f1_score(y_true, y_pred, average="binary"),
            "Cohen's Kappa": cohen_kappa_score(y_true, y_pred)
        }
        return metrics_dict

    @staticmethod
    def _calculate_specificity(y_true, y_pred):
        """
        Calculates specificity for predictions.
        """
        tn, fp, _, _ = confusion_matrix(y_true, y_pred).ravel()
        specificity_score = tn / (tn + fp) if (tn + fp) > 0 else 0
        return specificity_score
```

```python
# Define the Scikeras model
def residual_block(input_tensor, units, dropout_rate):
    x = Dense(units)(input_tensor)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = Dropout(dropout_rate)(x)

    x = Dense(units)(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    x = Add()([x, input_tensor])
    x = Activation("relu")(x)
    return x

def create_keras_model(input_dim):
    inputs = Input(shape=(input_dim,))
    x = Dense(128, activation="relu")(inputs)
    x = residual_block(x, 128, 0.2)
    x = residual_block(x, 128, 0.2)
    outputs = Dense(1, activation="sigmoid")(x)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

```python
# Define the Skorch model
class AttentionBlock(nn.Module):
    def __init__(self, feature_dim):
        super(AttentionBlock, self).__init__()
        self.attention_weights = nn.Parameter(torch.randn(feature_dim, requires_grad=True))
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, x):
        # Apply softmax to attention weights, then scale input features
        weights = self.softmax(self.attention_weights)
        return x * weights

class PyTorchNet(nn.Module):
    def __init__(self, num_features):
        super(PyTorchNet, self).__init__()
        self.fc1 = nn.Linear(num_features, 256)
        self.attention = AttentionBlock(256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 1)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = x.float()
        x = self.relu(self.fc1(x))
        x = self.attention(x)
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

# Hyperparameter Tuning

```python
# Define the cross validation method
cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
```

```python
# Instantiate the best model and feature selector
best_model = XGBClassifier()
best_feature_selector = RFECV(
    estimator=best_model,
    step=1,
    cv=cv_strategy,
    scoring="accuracy"
)

# Prepare the dataset
best_pipeline = Pipeline([
    ("imputer", IterativeImputer(random_state=42)),
    ("scaler", StandardScaler()),
    ("feature_selector", best_feature_selector),
    ("classifier", best_model)
])
```

```python
# Define the parameter grids
param_grid_genetic = {
    'classifier__max_depth': sklearn_genetic.space.Integer(3, 10),
    'classifier__learning_rate': sklearn_genetic.space.Continuous(0.01, 0.3),
    'classifier__n_estimators': sklearn_genetic.space.Integer(100, 1000),
    'classifier__gamma': sklearn_genetic.space.Continuous(0, 5),
    'classifier__min_child_weight': sklearn_genetic.space.Integer(1, 10),
    'classifier__subsample': sklearn_genetic.space.Continuous(0.5, 1),
    'classifier__colsample_bytree': sklearn_genetic.space.Continuous(0.5, 1),
    'classifier__colsample_bylevel': sklearn_genetic.space.Continuous(0.5, 1),
    'classifier__colsample_bynode': sklearn_genetic.space.Continuous(0.5, 1),
    'classifier__reg_alpha': sklearn_genetic.space.Continuous(0, 5),
    'classifier__reg_lambda': sklearn_genetic.space.Continuous(1, 5),
    'classifier__scale_pos_weight': sklearn_genetic.space.Continuous(1, 100),
    'classifier__max_delta_step': sklearn_genetic.space.Integer(0, 10)
}

param_grid_bayes = {
    'classifier__max_depth': skopt.space.Integer(3, 10),
    'classifier__learning_rate': skopt.space.Real(0.01, 0.3),
    'classifier__n_estimators': skopt.space.Integer(100, 1000),
    'classifier__gamma': skopt.space.Real(0, 5),
    'classifier__min_child_weight': skopt.space.Integer(1, 10),
    'classifier__subsample': skopt.space.Real(0.5, 1),
    'classifier__colsample_bytree': skopt.space.Real(0.5, 1),
    'classifier__colsample_bylevel': skopt.space.Real(0.5, 1),
    'classifier__colsample_bynode': skopt.space.Real(0.5, 1),
    'classifier__reg_alpha': skopt.space.Real(0, 5),
    'classifier__reg_lambda': skopt.space.Real(1, 5),
    'classifier__scale_pos_weight': skopt.space.Real(1, 100),
    'classifier__max_delta_step': skopt.space.Integer(0, 10)
}

param_grid_halving = {
    'classifier__max_depth': randint(3, 10),
    'classifier__learning_rate': uniform(0.01, 0.3 - 0.01),
    'classifier__n_estimators': randint(100, 1000),
    'classifier__gamma': uniform(0, 5),
    'classifier__min_child_weight': randint(1, 10),
    'classifier__subsample': uniform(0.5, 0.5),
    'classifier__colsample_bytree': uniform(0.5, 0.5),
    'classifier__colsample_bylevel': uniform(0.5, 0.5),
    'classifier__colsample_bynode': uniform(0.5, 0.5),
    'classifier__reg_alpha': uniform(0, 5),
    'classifier__reg_lambda': uniform(1, 4),
    'classifier__scale_pos_weight': uniform(1, 99),
    'classifier__max_delta_step': randint(0, 10)
}
```

```python
# Configure tuners
tuners = {
    "Genetic": GASearchCV(
        estimator=best_pipeline,
        param_grid=param_grid_genetic,
        cv=cv_strategy,
        scoring="accuracy",
        population_size=10,
        generations=5,
        crossover_probability=0.9,
        mutation_probability=0.05,
        tournament_size=3,
        elitism=True
    ),
    "Bayesian": BayesSearchCV(
        estimator=best_pipeline,
        search_spaces=param_grid_bayes,
        cv=cv_strategy,
        scoring="accuracy",
        n_iter=50
    ),  # Make sure to change np.int instances to np.int32 in the scikit-optimize.skopt.space.transformers.py source code
    "Halving Random": HalvingRandomSearchCV(
        estimator=best_pipeline,
        param_distributions=param_grid_halving,
        cv=cv_strategy,
        scoring="accuracy",
        n_candidates=50,
        factor=3
    )
}
```

```python
# Iterate through the tuners
for tuner_name, tuner in tuners.items():
    print(f"\nRunning {tuner_name} Search...")
    tuner.fit(X_train, y_train)
    print(f"Best parameters found by {tuner_name}: {tuner.best_params_}")
    print(f"Best score achieved by {tuner_name}: {tuner.best_score_}")
```
