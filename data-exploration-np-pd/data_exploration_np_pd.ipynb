## Continuous Columns

```python
# Continuous columns
continuous_data = data[['age', 'daily_rate', 'distance_from_home', 'hourly_rate', 'monthly_income',
                        'monthly_rate', 'percent_salary_hike', 'total_working_years',
                        'training_times_last_year', 'years_at_company', 'years_in_current_role',
                        'years_since_last_promotion', 'years_with_curr_manager']]

continuous_data.head()
```

## Ordinal Columns

```python
# Ordinal columns
ordinal_data = data[['age_group', 'business_travel', 'education', 'environment_satisfaction',
                     'job_involvement', 'job_level', 'job_satisfaction', 'salary_slab',
                     'performance_rating', 'relationship_satisfaction', 'stock_option_level',
                     'work_life_balance']]

ordinal_data.head()
```

```python
# Rename salary slab to income range
ordinal_data = ordinal_data.rename(columns={'salary_slab': 'income_range'})
```

```python
# Function to print value counts for each column
def display_value_counts(df):
    """
    Display value counts for each column in the DataFrame.
    """
    for column in df.columns:
        column_value_count = df[column].value_counts()
        print(f"{column.upper()} Value Counts:")
        display(column_value_count)
        print("\n")
```

```python
# Print value counts for ordinal columns
display_value_counts(ordinal_data)
```

```python
# Ordinal replacement operations
ordinal_data = ordinal_data.copy()

ordinal_data['age_group'] = ordinal_data['age_group'].replace({
    '18-25': 1,
    '26-35': 2,
    '36-45': 3,
    '46-55': 4,
    '55+': 5
})

ordinal_data['business_travel'] = ordinal_data['business_travel'].replace({
    'Non-Travel': 1,
    'Travel_Rarely': 2,
    'TravelRarely': 2,
    'Travel_Frequently': 3
})

ordinal_data['income_range'] = ordinal_data['income_range'].replace({
    'Upto 5k': 1,
    '5k-10k': 2,
    '10k-15k': 3,
    '15k+': 4
})

ordinal_data['stock_option_level'] = ordinal_data['stock_option_level'].replace({
    0: 1,
    1: 2,
    2: 3,
    3: 4
})
```

```python
# Only two values for performance rating - let's drop and make this a binary column later
ordinal_data = ordinal_data.drop(columns='performance_rating')
ordinal_data.head()
```

## Contingency Table

```python
# Contingency table of attrition by gender
contingency_table = pd.crosstab(index=merged_data['attrition'], columns=merged_data['gender'], rownames=['Attrition'], colnames=['Gender'])

print(f"Contingency Table of Attrition by Gender:")
contingency_table
```

```python
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Attrition by Gender')
plt.show()
```

```python
# Plot observed frequencies
contingency_table.plot(kind='bar', figsize=(6, 4), color=['pink', 'blue'])
plt.title('Observed Attrition by Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Observed frequencies
O = contingency_table.values

# Calculate expected frequencies
row_totals = np.sum(O, axis=1).reshape(-1, 1)
col_totals = np.sum(O, axis=0).reshape(1, -1)
grand_total = np.sum(O)
E = row_totals @ col_totals / grand_total

# Compute the chi-square statistic
chi2 = np.sum((O - E)**2 / E)

critical_value = 3.841

print(f"The Pearson Chi-Square statistic is {round(chi2, 3)} with df = 1")
if chi2 > critical_value:
    print(f"Statistical significance found. Reject the null hypothesis.")
else:
    print(f"No statistical significance. Cannot reject the null hypothesis.")
```

```python
# Manual odds ratio computation
odds_female = 87 / 504
odds_male = 151 / 738

OR_gender = odds_male / odds_female

print(f"Odds of Attrition for Males to Females:")
round(OR_gender, 3)
```

```python
# Natural logarithm of the odds ratio
log_OR = np.log(OR_gender)

# Standard error of the log OR
std_error = np.sqrt(1/151 + 1/738 + 1/87 + 1/504)

# Compute the 95% confidence intervals
lower_bound = np.exp(log_OR - 1.96 * std_error)
upper_bound = np.exp(log_OR + 1.96 * std_error)

print(f"95% Odds Ratio Confidence Interval:")
print((round(lower_bound, 4), round(upper_bound, 4)))
```
